### Vì data không có nhãn $9$ nên ta cần bỏ cột output là $9$ ra khỏi W 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import copy
```

```python
df_W = pd.DataFrame(data=np.genfromtxt("LogisticWeight.csv", delimiter=','),
                    columns=range(0, 25))
df_W = df_W.drop(columns=9)
df_W
```

```python
test = np.genfromtxt("Logistic_data_test.csv", delimiter=',')
X_test = test[:, 1:]
Y_test = test[:, 0:1]
```

### Đây mới là độ chính xác đúng trên tập test:

```python
lgt_vals = 1 / (1 + np.exp(-X_test @ df_W))
lgt_vals = lgt_vals.T
```

```python
np.mean(Y_test.reshape(Y_test.shape[0]) == lgt_vals.idxmax())
```

### Những hình có output sai:

```python
wrong = np.where(Y_test.reshape(Y_test.shape[0]) != lgt_vals.idxmax())[0]
```

```python
def display_image(img):
    plt.figure(figsize=(1, 1))
    img_copy = copy.copy(img.reshape((28, 28)))
    plt.imshow(img_copy, cmap='gray')
    plt.axis('off')
```

```python
for w in wrong:
    display_image(X_test[w, 1:])
```
